```python
# SEABORN DATA VISUALIZATION LIBRARY

#import warnings
#warnings.filterwarnings("ignore")
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.__version__)
```

```python
sns.get_dataset_names()
```

```python
dataset = sns.load_dataset("iris")
dataset
```

```python
# Line Graph
# lineplot()
sns.lineplot(x="sepal_length",y="sepal_width",data=dataset)
```

```python
# adding style to seaborn
plt.figure(figsize=(10,5))
sns.set_style("whitegrid",{"grid.color":"lightblue","grid.linestyle":"--"})# -- dashed : dotted
sns.lineplot(x="sepal_length",y="sepal_width",data=dataset)
plt.title("Line Plot In Seaborn",color="blue",fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
pokemon = pd.read_csv("https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv")
pokemon
```

```python
# scatterplot()
sns.scatterplot(x="Attack",y="Defense",data=pokemon)
plt.title("Scatterplot In Seaborn",color="blue",fontsize=10)
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
sns.scatterplot(x="Attack",y="Defense",hue="Generation",data=pokemon)
plt.title("Scatterplot In Seaborn",color="blue",fontsize=10)
plt.show()
```

```python
colors = ["r","y","b","g","c","m"]
sns.scatterplot(x="Attack",y="Defense",hue="Generation",palette=colors,data=pokemon)
plt.title("Scatterplot In Seaborn",color="blue",fontsize=10)
plt.show()
```

```python
# lmplot()
# used to generate a scatter plot with a regression line

colors = ["r","y","b","g","c","m"]
sns.lmplot(x="Attack",y="Defense",hue="Generation",palette=colors,data=pokemon)
plt.title("lmplot In Seaborn")
plt.show()
```

```python
# countplot()
# used to generate a bar graph
plt.figure(figsize=(15,5))
sns.countplot(x="Type 1",data=pokemon)
plt.title("Countplot In Seaborn")
plt.show()
```

```python
pkmn_type_colors = ['#3ae40c',  # Grass
                    '#e52106',  # Fire
                    '#0AE3D6',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#ECF709',  # Dragon
                    '#7A786C',  # Dark
                    '#D3D1C0',  # Steel
                    '#91E8EE',  # Flying
                   ]
```
