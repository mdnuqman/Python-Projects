```python
# Assignment 1 String Method

sentence = input("Write a sentence : ")

choice = int(input("""Let's start!

             1.capitalize 2.casefold
             3.count      4.endswith
             5.find       6.islower
             7.isupper    8.lower
             9.upper      10.startswith
             11.swapcase  12.replace

             Pick a number : """))

if choice == 1:
    print(sentence.capitalize())
    print("\nDefinition : The capitalize() method returns a string where the first character is upper case, and the rest is lower case.")
elif choice == 2:
    print(sentence.casefold())
    print("\nDefinition : The casefold() method returns a string where all the characters are lower case.")
elif choice == 3:
    word = input("Enter a word : ")
    print(sentence.count(word))
    print("\nDefinition : The count() method returns the number of times a specified value appears in the string.")
elif choice == 4:
    end = input("Your sentence ends with : ")
    print(sentence.endswith(end))
    print("\nDefinition : The endswith() method returns True if the string ends with the specified value, otherwise False.")
elif choice == 5:
    search = input("Find what : ")
    result = sentence.find(search)
    if result == -1:
        print("Not found")
    else:
        print(result)
    print("\nDefinition : The find() method finds the first occurrence of the specified value.")
elif choice == 6:
    print(sentence.islower())
    print("\nDefinition : The islower() method returns True if all the characters are in lower case, otherwise False.")
elif choice == 7:
    print(sentence.isupper())
    print("\nDefinition : The isupper() method returns True if all the characters are in upper case, otherwise False.")
elif choice == 8:
    print(sentence.lower())
    print("\nDefinition : The lower() method returns a string where all characters are lower case.")
elif choice == 9:
    print(sentence.upper())
    print("\nDefinition : The upper() method returns a string where all characters are in upper case.")
elif choice == 10:
    input_start = input("Your sentence starts with : ")
    print(sentence.startswith(input_start))
    print("\nDefinition : The startswith() method returns True if the string starts with the specified value, otherwise False.")
elif choice == 11:
    print(sentence.swapcase())
    print("\nDefinition : The swapcase() method returns a string where all the upper case letters are lower case and vice versa.")
elif choice == 12:
    replace_word = input("What word do you want to replace : ")
    replace_with = input("What word do you want to replace with : ")
    print(sentence.replace(replace_word,replace_with))
    print("\nDefinition : The replace() method replaces a specified phrase with another specified phrase.")
else:
    print("Only 1-12")
```
