```python
# TYPE CASTING
# IT'S A PROCESS OF CONVERTING ONE DATA TYPE TO ANOTHER DATA TYPE

# automatic type conversion    IMPLICIT
# lower data type will be converted to higher data type automatically

a = 10
b = 45.03
res = a + b     # 10.00 + 45.03
print(res)
```

```python
# manual type conversion  EXPLICIT
# performed by developers

a = 10
b = 45.03
res = a + int(b)   # 10 + 45
print(res)


# we have a floatVar which needs to be converted into intVar
#  int(floatvar)      int(23.33) --> 23

# we have an intVar which needs to be converted into floatVar
#  float(intVar)     float(10)  --> 10.0

# we have a intVar which needs to be converted into strVar
#  str(intVar)         str(24)   --> '24'

# we have a floatVar which needs to be converted into strVar
#  str(floatVar)         str(24.334)   --> '24.334'

# converting bool into int and float

choice = True
print(int(choice))
print(float(choice))

choice = False
print(int(choice))
print(float(choice))

name = "89.21"
print(float(name))
```

```python
# input() function will get the value in a form of STRING

val1 = input('Enter The First Value : ')      # '20'
val2 = input('Enter The Second Value : ')     # '26'
res = val1 + val2                             # res = '20'+'26'
print(f"After Addition {val1} + {val2} Result Will Be {res}")
```

```python
val1 = int(input('Enter The First Value : '))       # 20
val2 = int(input('Enter The Second Value : '))      # 26
res = val1 + val2                                   # res = 20+26
print(f"After Addition {val1} + {val2} Result Will Be {res}")
```

```python
# BMI Exercise
print("This is the BMI calculator.")
name = input("Enter your name: ")
weight_kg = float(input("Enter your weight in kg: "))
height_cm = float(input("Enter your height in cm: "))

height_m = height_cm/100

BMI = weight_kg/(height_m * height_m)

print(f"Hi {name}, your BMI is {BMI}.")
```

```python
# Operators
# Operators are used to perform operations on Variables

# a.) ARITHMETIC OPERATORS
# ARE USED TO PERFORM MATHEMATICAL OPERATIONS ON VARIABLES
#  +  -   *    /   //  %  **

print(10+5)
print(10-5)
print(10*5)
print(10/5)   # IN PYTHON DIVISION OPERATOR / USED TO RETURN VALUE IN FLOAT
print(10//5)  # IN PYTHON FLOOR DIVISION OPERATOR // USED TO RETURN VALUE IN INT
print(10%3)   # modulus used to give remainder
print(2**8)   # 2 raise to the power of 8   ** Exponent
print(72**72)
```

```python
# b.) ASSIGNMENT OPERATOR
# used to assign a value to a variable

a = 90      # a as 90  or  a is 90
print(a)
```

```python
# c.) COMPARISON OPERATORS
# <  >   ==  !=   <=   >=
# USED TO RETURN A BOOLEAN VALUE  [True / False]

print(10<5)
print(10>5)
print(98==98)
print(100!=1000)
print(100!=100)

age = 18
print(age>=18)  # age should be equal to 18 or age should be greater than 18
```

```python
# d.) LOGICAL OPERATORS
#  works with multiple conditions
#  and  or  not
#  USED TO RETURN A BOOLEAN VALUE  [True / False]

# and
# and requires more than one condition
# and will return TRUE when both or all the conditions are TRUE

print(100==100 and 200==200 and 300==300)
#       True          True        True

print(100==100 and 200==2000)
#       True          False
```

```python
# or
# or requires more than one condition
# or will return TRUE when both or all the conditions are TRUE
# or will also returns True if atleast any condition is TRUE
# or will return FALSE when both or all the conditions are FALSE

print(100==100 or 200==200 or 300==300)
#       True          True        True

print(100==100 or 200==2000)
#       True          False

print(100==1000 or 200==2000)
#       False          False
```

```python
# e.) Membership Operators
#     in   not in

# in
# used to return True
# if the specified value exists in a Sequence

print("P" in "Python")
print("o" in "Python")
print("John" in ["John","Smith","Alex"])
print("x" in "Python")
print("p" in "Python")
```
