```python
# Exception handling in Python is a mechanism that allows you to handle runtime errors,
# or exceptions, in a controlled way, rather than letting your program crash unexpectedly.
# Python provides a way to catch errors, respond to them, and allow your program to continue
# running smoothly even if something goes wrong.
# try except finally
# try : keep that block of code inside try , which might give you an error
# except : catch that error
```

```python
try:
  val1 = int(input("Enter first number : "))
  val2 = int(input("Enter second number : "))
  result = val1 / val2
  print(f"After division result will be {result}")
except:
  print("Something went wrong")
  print("Make sure to provide only integer input or second number cant be zero")
print("Rest of the code")
```

```python
# Using An Exception Class
# if you need to know the reason for an error use Exception Class

try:
  val1 = int(input("Enter first number : "))
  val2 = int(input("Enter second number : "))
  result = val1 / val2
  print(f"After division result will be {result}")
except Exception as e:
  print("Something went wrong")
  print(f"Error cause by : {e}")
print("Rest of the code")
```

```python
# case no.3 : when there is an exception but not handelled properly

try:
  val1 = int(input("Enter first value : "))
  val2 = int(input("Enter second value : "))
  result = val1 / val2
  print(f"After division result will be {result:.2f}")
except ZeroDivisionError as e: # Changed from ImportError to ZeroDivisionError
  print("Something went wrong")
  print(f"Error cause by : {e}")
except ValueError as e: # Added to handle invalid integer input
  print("Something went wrong with input")
  print(f"Error cause by : {e}")
except Exception as e: # Added to catch any other unexpected errors
  print("An unexpected error occurred")
  print(f"Error cause by : {e}")
print("Rest of the code")
```

```python
try:
    import non_existent_module_xyz
    print("Module imported successfully!")
except ImportError as e:
    print(f"Caught an ImportError: {e}")
    print("This usually means the module is not found or not installed.")
except Exception as e:
    print(f"Caught an unexpected error: {e}")

print("Program continues after attempting to import a non-existent module.")
```
