```python
# for loop
# for loop is used to iterate over a sequence [collection of data values]

# syntax
# for iterate_variable in Sequence:
#     block of code
price = 12, 312, 12, 41, 41, 53, 531, 513, 13, 32
tax = 2.05

for i in price:
  print(i + tax)
```

```python
# range()
# it's a built-in function and it is is used to generate
# a sequence of Integer Numbers

# a.) range(x)
# it is is used to generate a sequence of Integer Numbers
# starting from Zero till x-1

print(list(range(10)))

# b.) range(x , y)
# it is is used to generate a sequence of Integer Numbers
# starting from x till y-1

print(list(range(1,11)))

# c.) range(x , y , stepCount/Interval)
# it is is used to generate a sequence of Integer Numbers
# starting from x till y-1 based on stepCount

print(list(range(1,11,3)))

# print all teh even numbers till 100 using range

print(list(range(0,101,2)))
```

```python
num = int(input("Number : "))

for i in range(1,11):
  print(f"{num} x {i} = {num*i}")
```

```python
# NESTED FOR LOOP

# OUTER LOOP WILL WAIT UNTIL  INNER LOOP FINISHES THE EXECUTION

for c in ["red" , "blue" , "green"]:    # OUTER LOOP
  for v in ["car" , "bike" , "bus"]:    # INNER LOOP
    print(c , v)
```

```python
# while loop
# while loop repeats a block of code
# until a certain condition met

# syntax

# initialization
# while condition:
#  a block of code
#  increment / decrement

# While checks the condition first
# if the condition is TRUE while will execute the block of code
# after executing the block of code while will check the condition again


i = 1
while i<=3:    # 1<=3 (t)           2<=3(t)             3<=3(t)         4<=3(F)
  print(i)     # 1                  2                   3
  i += 1       # i = 1 + 1 (2)      i = 2 + 1 (3)       i = 3 + 1 (4)
```

```python
# print mathemetical table of a number using while loop
num = int(input('Enter Any Number : '))
i = 1
while i<=10:
  print(f"{num} x {i} = {num*i}")
  i += 1
```

```python
# Use while loop when the number of iterations are not known

num = 121
temp = num
sum = 0

while num!=0:           # 1st Iteration  121!=0(t)   2nd Iteration  12!=0(t)   3rd Iteration 1!=0(t)   4th Iteration 0!=0 (f)
  rem = num % 10        #  rem = 1                    rem = 2                    rem = 1
  sum = sum*10 + rem    #  sum = 1                    sum = 12                   sum = 121
  num = num // 10       #  num = 12                   num = 1                    num = 0

if temp == sum:
  print('Palindrome Number')
else:
  print('Not A Palindrome')
```

```python
# break and continue statement

# break statement
# it is used to terminate the execution of a LOOP

# break and continue should be defined inside decision making

for i in range(1,11):
  if i == 5:
    break
  print(i)
```

```python
# continue statement
# it is used to skip the current execution of a LOOP

for i in range(1,11):
  if i == 5 or i == 9:
    continue
  print(i)
```
