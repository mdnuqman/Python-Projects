```python
# if statement
# if will check the condition First
# if the condition is TRUE the only it will allow python
# interpreter to execute the code

# Syntax
# if condition:
# block of code

age = 21
if age>=18:
  print("Eligible For A Driving License")
```

```python
num = int(input("Enter a number : "))
if num%2==0:
  print("Even")
else:
  print("Odd")
```

```python
a = 19
b = 45
c = 27
d = 141
e = 67

largest = 0

if a>largest:       # 19>0(t)
  largest = a       # largest = 19
if b>largest:       # 45>19(t)
  largest = b       #largest = 45
if c>largest:       #27>45(f)
  largest = c       #largest = 45
if d>largest:       #141>45(t)
  largest = d       #largest = 141
if e>largest:       #141>67(f)
  largest = e       #largest = 141

print(f"Largest is : {largest}")
```

```python
# if else
age = 18
if age >=18:
  print("Welcome user")
  print("Game is loading...")
  print("Please wait")
else:
  print("Age Restriction.")
  print("Age should be ABOVE or EQUAL to 18")
```

```python
# One liner if else
# ternary operator

# if_condition_is_true if condition else value_if_condition_is_false

num = int(input("Enter Any Number : "))
result = "Even number" if num%2==0 else "Odd number"
print(result)
```

```python
# if elif else
# working with multiple conditions

# at a time only one block of code will be executed

num = int(input("Enter any number : "))
if  num<0:
  print("Negative Number")
elif num==0:
  print("Zero")
else:
  print("Positive Number")
```

```python
marks = int(input("Enter your marks : "))
if marks>100 or marks<0:
  print("Invalid Marks")
elif marks>=85:
  print("Grade A")
elif marks>=75:
  print("Grade B")
elif marks>=60:
  print("Grade C")
elif marks>=40:
  print("Grade D")
elif marks>=0:
  print("Grade F")
```

```python
# Chained Comparison
# help us to define a range
# Lower Limit <=  value   <=  Upper Limit

print("Grading Allotment Application")
marks = int(input("Enter Your Marks :"))   # 83

# here 85 <= marks <= 100 this expression means
#  a.) the value is less than or equal to 100    c1     is 93 <=100 (t)
#  b.) 85 means that I am less the define value  c2     is 85 is less than 93 (t)

if 85 <= marks <= 100:
    print('Grade A')
elif 75 <= marks <= 84:
    print('Grade B')
elif 60 <= marks <= 74:
    print('Grade C')
elif 40 <= marks <= 59:
    print("Grade D")
elif 0 <= marks <= 39:
    print("Grade F")
else:
    print("Invalid Marks")
```

```python
name = input("What is your name? : ")
weight = float(input("What is your weight(kg)? : "))
height = float(input("What is your height(cm)? : "))

metre = height/100

bmi = round(weight/(metre**2),1)

if bmi<18.5:
  print(f"\n{name} you are underweight 😌")
  print(f"Your BMI is {bmi}")
elif 18.5<=bmi<=24.9:
  print(f"\n{name} you are normal weight 😁")
  print(f"Your BMI is {bmi}")
elif 25<=bmi<=29.9:
  print(f"\n{name} you are overweight 😔")
  print(f"Your BMI is {bmi}")
else:
  print(f"\n{name} you are obese 😱
  ")
  print(f"Your BMI is {bmi}")
```
