```python
# Q1. Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

# Output
# No. Of Positive : 4
# No. Of Negative : 3

a = [10, -20, 30, -40, 50, -60, 0]

positive_count = 0
negative_count = 0

for num in a:
  if num > 0:
    positive_count += 1
  elif num < 0:
    negative_count += 1

print("No. Of Positive :", positive_count)
print("No. Of Negative :", negative_count)
```

```python
# Q2. Using List Comprehension. Write a Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

a = [10, -20, 30, -40, 50, -60, 0]

positive_count = len([num for num in a if num>0])
negative_count = len([num for num in a if num<0])

print("No. Of Positive :", positive_count)
print("No. Of Negative :", negative_count)
```

```python
# Q3. Remove Multiple Elements from List in Python
# a = [10, 20, 30, 40, 50, 60, 70]
# remove = [20, 40, 60]
# Resulting list = [10, 30, 50, 70]

a = [10, 20, 30, 40, 50, 60, 70]
remove = [20, 40, 60]

result = [num for num in a if num not in remove]
print(result)
```

```python
# Q4. Remove duplicates from list in Python
# a = [1, 2, 2, 3, 4, 4, 5]

# Output : [1,2,3,4,5]

a = [1, 2, 2, 3, 4, 4, 5]

result = list(set(a))
print(result)
```

```python
# Q6. Key with Maximum Unique Values - Python
# dict1 = {"key1": [5, 7, 5, 4, 5], "key2": [6, 7, 4, 3, 3], "key3": [9, 9, 6, 5, 5]}
# output : key2

dict1 = {"key1": [5, 7, 5, 4, 5], "key2": [6, 7, 4, 3, 3], "key3": [9, 9, 6, 5, 5]}

max_unique_count = -1
max_unique_key = None

for key, value in dict1.items():
  unique_values = len(set(value))
  if unique_values > max_unique_count:
    max_unique_count = unique_values
    max_unique_key = key

print(max_unique_key)
```

```python
# Q7. Story: An online store records daily product sales.
# Convert this list into total sales per product.
sales = [
    ("apple", 3),
    ("banana", 2),
    ("apple", 5),
    ("banana", 1),
    ("orange", 4)
]

# Output : {'apple': 8, 'banana': 3, 'orange': 4}

total_sales = {}
for product, quantity in sales:
    total_sales[product] = total_sales.get(product, 0) + quantity

print(total_sales)
```

```python
# Q8. Undo Feature in Text Editor
# A text editor supports an undo operation.
# Given operations:
# actions = ["type A", "type B", "delete", "type C"]
# Output : ['type A', 'type C']

# actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]
# Output : ['type A', 'type C', 'type A']

# Example 1
actions1 = ["type A", "type B", "delete", "type C"]
stack1 = []
for action in actions1:
  if action.startswith("type"):
    stack1.append(action)
  elif action == "delete" and stack1:
    stack1.pop()

print(stack1)

actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]

stack2 = []
for action in actions:
  if action.startswith("type"):
    stack2.append(action)
  elif action == "delete" and stack2:
    stack2.pop()

print(stack2)
```

```python
# Q9. Print the index number of those Two boxes whose weights sum to a target.
# weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
# target = 9

# Output
'''
0 1
4 5
0 6
4 7
0 8
8 9
'''
weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
target = 9

for i in range(len(weights)):
  for j in range(i , len(weights)):
    if weights[i] + weights[j] == target:
      print(i, j)

```

```python
# Q10. Online Store Price Update (map + lambda)
# An online store increases all product prices by 10%.
# prices = [100, 200, 300, 400]
# output [110, 220, 330, 440]

prices = [100, 200, 300, 400]

new_prices = list(map(lambda x: int((x+(x*0.1))),prices))
print(new_prices)
```

```python
# Q11. Employee Bonus System (map + if–else)
# Employees earning ≥ 50,000 get a 20% bonus, others get 10%.
# salaries = [40000, 60000, 50000, 30000]

salaries = [40000, 60000, 50000, 30000]
list(map(lambda x: x+int(x*0.2) if x>=50000 else x+int(x*0.1),salaries))
```

```python
# Q12. Order Filtering System (filter + multiple conditions)
# An e-commerce system wants orders above ₹1000 and not cancelled.

orders = [
    {"amount": 1200, "status": "completed"},
    {"amount": 800, "status": "completed"},
    {"amount": 1500, "status": "cancelled"},
    {"amount": 2000, "status": "completed"}
]

after_order = list(filter(lambda x: x["amount"]>1000 and x["status"]=="completed",orders))

print(after_order)
```

```python
# Q13. Data Cleaning Pipeline (map + filter + lambda)
# From a data pipeline:
#  a.)Remove negative numbers
#  b.) Square remaining numbers
#  c.) Keep only results > 50

# data = [-4, -1, 2, 4, 8]
# Output : [64]

data = [-4, -1, 2, 4, 8]

after_cleaning = list(filter(lambda x: x>0,data))
after_cleaning2 = list(map(lambda x:x**2,after_cleaning))
after_cleaning3 = list(filter(lambda x: x>50,after_cleaning2))

print(after_cleaning3)
```
