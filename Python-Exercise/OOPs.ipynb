```python
# OOPS
# OOP stands for Object-Oriented Programming.
# Object-oriented programming (OOP) is a programming paradigm based on the
# concept of "objects"
# Procedural programming is about writing procedures or methods that perform operations
# on the data, while object-oriented programming is about creating objects
# that contain both data and methods.
# Object-oriented programming has several advantages over procedural programming:
# a.) OOP is faster and easier to execute
# b.) OOP provides a clear structure for the programs
# c.) OOP helps to keep the Java/Python code DRY "Don't Repeat Yourself", and makes the code easier
#     to maintain, modify and debug
# d.) OOP makes it possible to create full reusable applications with less code and
#     shorter development time
# Class
# A class is a blueprint that defines the variables and the methods common to all objects.
# a class is a collection of methods and variables

# Object
# An object is called an instance of a class.

# Self Keyword
# The self parameter is a reference to the current instance of the class, and is used to
# access variables that belongs to the class.
# The first parameter of each method in a class must be the self ,
# which refers to the calling object. However, you can give any name to the first
# parameter, not necessarily self.
# __init__ in Python
#  Constructors are used to initializing the object’s state. The task of constructors is
# to initialize(assign values) to the data members of the class when
# an object of the class is created.
# The __init__() function is called automatically every time the class is being used to create
# a new object.
```

```python
# class is a collection
# a class is a collection of method/functions and variables
# in python a class can be defined by using a class keyword

class myClass:

  """
  This is my first class in python.
  I have just started learning OOPs
  """

  a = 10      # member variable
  b = 20

  def func1():    # member function
    print("I am the method of a class")

  def func2():
    print("I am the second method of a class")

# we can not access class variable and functions directly
# we need to call these variable and function with the help of the class

print (myClass.a)# . dot is a member access operator
myClass.func1()
```

```python
class myClass:

  """
  This is my first class in python.
  I have just started learning OOPs
  """
  a = 10      # member variable
  b = 20

  def func1():    # member function
    print("I am the method of a class")

  def func2():
    print("I am the second method of a class")

# In OOPs in order to access class variables and functions
# we need to define the Object of a class

# syntax for defining the object
# objname = className()

# object can access the class variable and functions
# in python a class can have a number of objects

obj1 = myClass()    # creating an object of a class
obj2 = myClass()    # creating an another object of class

print(obj1.a)
print(obj2.b)
```

```python
# Self keyword

#   Important Rule
# whenever you are defining any function inside the class
# the parameter of that function needs to be SELF

# SELF REPRESENT THE CURRRET CALLING OBJECT

class myClass:

  a = 10
  b = 20

  def func1(self):                          # 2. self value will be replaced by obj1 func1(obj1) func1(obj2)
    print("I am the method of a class")

  def func2(self):
    print("I am the second method of a class")

obj1 = myClass()
obj2 = myClass()

obj1.func1()
obj2.func1()
```

```python
# Design an application for me using Procedural Programming
# User Registeration System

# Task1
# Take 3 information from 3 users
# user1      user2    user3
# name       name     name
# email      email    email
# mobno      mobno    mobno

# Task2
# Print all this information

# Task3
# Print only the second user information

# Task4
# Print the email and mobno of 3rd User
user = []
user_data = 3

for i in range(user_data):
  print(f"\nEnter the user {i+1} data")
  name = input("Enter your name: ")
  email = input("Enter your email: ")
  mobno = input("Enter your mobno: ")
  user.append([name,email,mobno])

print("\n All User Information ")
for i, user_info in enumerate(user, 1):
  print(f"User {i}: Name : {user_info[0]}, Email : {user_info[1]}, Mobile : {user_info[2]}")

print("\n Second User Information ")
second_user = user[1]
print(f"Name : {second_user[0]}, Email : {second_user[1]}, Mobile : {second_user[2]}")

print("\n Third User's Email and Mobile Number ")
third_user = user[2]
print(f"Email : {third_user[1]}, Mobile : {third_user[2]}")
```

```python
class userInfo:

    def __init__(self,n,e,mn):    # us3  , n = Smith  ,   e = sm@gmail.com  , mn = 60690231
        self.name = n
        self.email = e
        self.mobno = mn

    def printDetails(self):                            # printDetails(us1)                       printDetails(us2)
        print(self.name,self.email,self.mobno)         #   us1.name us1.email us1.mobno           us2.name us2.email us2.mobno
us1 = userInfo("John","j@gmail.com",60987654)  # __init__ calls itself and this information ("John","j@gmail.com",60987654) will be passed to it automatically
us2 = userInfo("Alex","al@gmail.com",60453213) # __init__ calls itself and this information ("Alex","al@gmail.com",60453213) will be passed to it automatically
us3 = userInfo("Smith","sm@gmail.com",60690231) # __init__ calls itself and this information ("Smith","sm@gmail.com",60690231) will be passed to it automatically

print("====Printing All Information====")
us1.printDetails()
us2.printDetails()
us3.printDetails()

print("\n====Printing Second User Information====")
us2.printDetails()

print("\n====Printing Email and MobNo Of 3rd User====")
print(us3.email)
print(us3.mobno)
```
