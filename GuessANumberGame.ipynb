```python
import random

def play_guess_number_game():
    guessAttempt = 1

    print('🎈 Guess A Number Game')
    print('Enter Your Name')
    name = input()
    print(f'Welcome User {name} 😎')
    print('I am thinking of a number between 1 and 13 🧐. Please Wait.....')
    randomNum = random.randint(1,13)
    print('Number Is Generated ✅ Guess The Number In 4 Attempts!')
    while guessAttempt <= 4:
        print(f'Attempt{guessAttempt} Take A Guess.')
        try:
            guess = int(input())
        except ValueError:
            print('Invalid input. Please enter a number.')
            continue

        if guess == randomNum:
            print('🎉 Congratulation. Correct Guess 🎉')
            print(f'You have guess the number in {guessAttempt} attempt.')
            break

        if guess > randomNum:
            print('Your Guess Is Too HIGH ⬆️ Think Of A Smaller Number')

        if guess < randomNum:
            print('Your Guess Is Too LOW ⬇️ Think Of A Bigger Number')

        guessAttempt = guessAttempt + 1

    if guessAttempt > 4 or guess != randomNum:
        print('❌ GAME OVER! ❌')
        print(f'The Correct Number Was {randomNum} ❇️')

play_again = True
while play_again:
    play_guess_number_game()
    response = input('Do you want to play again? (yes/no): ').lower()
    if response != 'yes':
        play_again = False
    print('Thanks for playing! 👋')
```
